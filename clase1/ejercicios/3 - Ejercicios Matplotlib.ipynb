Ejercicio 2: Desarrolle una gráfica de linea usando el módulo pyplot de Matplotlib para producir la función:

$$ f(x) = x^2 |sen(x)| - 4 $$

Sobre el dominio $ 0 \leq x \leq 4 $. Asegurese de que los ejes y la curva esté correctamente etiquetadas. Agregue un titulo. 

```python
import numpy as np
# linspace(start, stop, num_points)
x = np.linspace(0,4,100)
f = ((x ** 2) * np.abs(np.sin(x))) - 4
plt.plot(x, f, c="r", linestyle="-");
# Agrego una grilla
plt.grid()
# Nombre del eje x
plt.xlabel("x")
# Nombre del eje y
plt.ylabel("f(x)")
# Titulo del grafico
plt.title("Gráfica de x^2 |sen(x)| - 4")
plt.show()
```

Ejercicio 5:

Cree un gráfico en estructura de dos por dos usando **figure** para las siguientes funciones hiperbólicas en el dominio $ -1 \leq x \leq 1$

| Estructura del grafico | |
| ---------------------  | -- |
| $senh(x) \; vs \; x$ | $cosh(x) \; vs \; x$ |
| $tanh(x) \; vs \; x$ | $sech(x) \; vs \; x$ |

```python
x = np.linspace(-1,1,100)
senh_x = np.sinh(x)
cosh_x = np.cosh(x)
tanh_x = np.tanh(x)
# No hay funcion sech, hay que hacerla por identidades
sech_x = 1 / np.cosh(x)
fig = plt.figure(figsize=(12, 12))
# senh(x) vs x en parcela 2
ax1 = fig.add_subplot(221)
ax1.plot(x, senh_x, c="r", linestyle="-", label="senh(x)")
ax1.grid()
ax1.set_xlabel("x")
ax1.set_ylabel("senh(x)")
ax1.set_title("senh(x) vs x")
# senh(x) vs x en parcela 1
ax2 = fig.add_subplot(222)
ax2.plot(x, cosh_x, c="b", linestyle="-", label="cosh(x)")
ax2.grid()
ax2.set_xlabel("x")
ax2.set_ylabel("cosh(x)")
ax2.set_title("cosh(x) vs x")
# tanh(x) vs x en parcela 3
ax3 = fig.add_subplot(223)
ax3.plot(x, tanh_x, c="g", linestyle="-", label="tanh(x)")
ax3.grid()
ax3.set_xlabel("x")
ax3.set_ylabel("tanh(x)")
ax3.set_title("tanh(x) vs x")
# sech(x) vs x en parcela 4
ax4 = fig.add_subplot(224)
ax4.plot(x, sech_x, c="m", linestyle="-", label="sech(x)")
ax4.grid()
ax4.set_xlabel("x")
ax4.set_ylabel("sech(x)")
ax4.set_title("sech(x) vs x")
# Espacios entre graficos dentro de la figura de forma automatica
fig.tight_layout();
plt.show()
```
